**Columna Price** - La columna tiene dos tipos de datos. Str y float. Buscamos pasar todos los datos a float.  

```python
steam_games["price_number"] = pd.to_numeric(steam_games["price"], errors="coerce")
steam_games["price_number"] = pd.to_numeric(steam_games["price_number"], errors="coerce")
steam_games["price_number"] = steam_games["price_number"].fillna(0)
```

**Release date**- Creamos una nueva columna que indique año de lanzamiento

```python
invalid_dates = steam_games["release_date"][~steam_games["release_date"].str.match(r"\d{4}-\d{2}-\d{2}", na=False, case=False, flags=0)]
df_invalid_dates = pd.DataFrame({"invalid_dates": invalid_dates})
df_invalid_dates
```

```python
#Cambiamos el formato
steam_games["release_date"] = pd.to_datetime(steam_games["release_date"], errors="coerce", format='%Y-%m-%d', exact=False)
mean_release_date = steam_games["release_date"].mean()
steam_games["release_date_filled"] = steam_games["release_date"].fillna(mean_release_date)
steam_games["release_year"] = steam_games["release_date_filled"].dt.year
steam_games
```

**Genres**- Analizamos la variable genres y tags y la fusionamos ya que detallan lo mismo

```python
steam_games["genres"].isnull().sum()
```

```python
generos = []
for genres_list in steam_games["genres"].dropna():
    for genre in genres_list:
        if genre not in generos:
            generos.append(genre)
generos
```

```python
steam_games["tags"]=steam_games["tags"].dropna().apply(lambda lista: [g for g in lista if g in generos])
```

```python
steam_games["genres"].fillna(steam_games["tags"], inplace=True)
```

```python
steam_games["genres"].isnull().sum() #Chequeamos si aún encontramos registros nulos#
```

```python
steam_games = steam_games.drop("tags", axis=1) #Eliminamos la columna tags ya que no se utilizará"
steam_games
```

```python
# Se extrae el primer elemento de "genres" y se crea la columna "main_genre"
steam_games["main_genre"] = steam_games["genres"].apply(lambda x: x[0] if isinstance(x, list) and x else None)

# CSe crea un nuevo dataframe
steam_games_unique_genre = steam_games.drop_duplicates(["app_name", "title", "id", "developer", "price_number", "release_year"])
steam_games_unique_genre = steam_games_unique_genre.drop("genres", axis=1)

# Se muestra el nuevo dataframe creado
steam_games_unique_genre
```

```python
steam_games_unique_genre["main_genre"].isnull().sum()
```

```python
steam_games_unique_genre["main_genre"] = steam_games_unique_genre["main_genre"].fillna("Sin Dato")
steam_games_unique_genre["main_genre"].isnull().sum()
```

**Corroboramos las columnas App_name, title y developer**

```python
columnas_a_llenar = ["app_name", "title"]
steam_games_unique_genre[columnas_a_llenar] = steam_games_unique_genre[columnas_a_llenar].fillna("Sin datos")
```

```python
igualdad = (steam_games_unique_genre["app_name"] == steam_games_unique_genre["title"]).all()

if igualdad: #Definiremos si las columnas son lo mismo#
    print("Las columnas 'app_name' y 'title' son iguales en todos los registros.")
else:
    print("Las columnas 'app_name' y 'title' no son iguales en todos los registros.")
```

```python
steam_games_unique_genre["igualdad"] = steam_games_unique_genre["app_name"] == steam_games_unique_genre["title"]
filas_sin_igualdad = steam_games_unique_genre[steam_games_unique_genre["igualdad"] == False]
filas_sin_igualdad
```

```python
steam_games_unique_genre.dropna(subset=["developer"], inplace= True) #Borramos los nulos en developer#
```
