```python
user_items = "users_items.json.gz"
filas=[]
with gzip.open(user_items, 'rt', encoding='MacRoman') as archivo:
    for line in archivo.readlines():
        filas.append(ast.literal_eval(line))
#Transformamos a un dataframe#
user_item = pd.DataFrame(filas)
```

```python
user_item
```

```python
user_item.info()
```

```python
#Chequeamos los valores nulos#
user_item.isnull().sum()
```

```python
user_item.columns
```

Se puede apreciar que la variable items es una variable con datos anidados, a√∫n as√≠ se procede a chequearla

```python
user_item["items"][20]
```

Desanidamos la columna items con explode

```python
user_item = user_item.explode("items").reset_index(drop=True)
user_item
```

```python
user_item_normalizado = pd.json_normalize(user_item["items"])
user_item_normalizado
```

```python
user_item_final = pd.concat([user_item,user_item_normalizado], axis=1)
user_item_final
```

```python
user_item_final.columns
```

Realizado este proceso, eliminamos la columna items con datos anidados.

```python
user_item_final = user_item_final.drop(columns=["items"])
user_item_final
```

Eliminamos columnas innecesarias, "user_url", "steam_id" y "playtime_2weeks"

```python
user_item_final = user_item_final.drop(columns=["user_url","playtime_2weeks","steam_id"])
user_item_final
```

```python
user_item_final.info()
```

Chequeamos duplicados

```python
duplicados = user_item_final.duplicated().sum()
duplicados
```

```python
duplicados_mask = user_item_final.duplicated(keep=False)
filas_duplicadas = user_item_final[duplicados_mask]
filas_duplicadas
```

Los duplicados corresponden al mismo usuario pero sobre diferentes juegos, se conservan.

Finalmente, se carga el dataset trabajado.

```python
from google.colab import files


user_item_final.to_csv('User_items_limpio.csv', index=False, encoding='utf-8')


files.download('User_items_limpio.csv')
```
